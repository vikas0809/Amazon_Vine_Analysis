```python
import pandas as pd
```

```python
df=pd.read_csv("Data/vine_table.csv")
```

```python
df
```

```python
step1_filter_df=df.loc[df["total_votes"]>=20]
step1_filter_df
```

```python
step1_filter_df["vote_ratio"]=round(step1_filter_df["helpful_votes"]/step1_filter_df["total_votes"]*100,2)
```

```python
step1_filter_df
```

```python
step2_filter_df=step1_filter_df.loc[step1_filter_df["vote_ratio"]>=50.00]
step2_filter_df
```

```python
total_five_star_reviews_df=step2_filter_df.loc[step2_filter_df["star_rating"]==5.0]
total_five_star_reviews=len(total_five_star_reviews_df)
total_five_star_reviews
```

```python
paid_reviews_df=step2_filter_df.loc[step2_filter_df["vine"]=="Y"]
paid_reviews_df
```

```python
paid_five_star_reviews_df=paid_reviews_df.loc[paid_reviews_df["star_rating"]==5.0]
paid_five_star_reviews=len(paid_five_star_reviews_df)
paid_five_star_reviews
```

```python
paid_reviews_ratio=round(paid_five_star_reviews/len(paid_reviews_df)*100,2)
paid_reviews_ratio
```

```python
unpaid_reviews_df=step2_filter_df.loc[step2_filter_df["vine"]=="N"]
unpaid_reviews_df
```

```python
unpaid_five_star_reviews_df=unpaid_reviews_df.loc[unpaid_reviews_df["star_rating"]==5.0]
unpaid_five_star_reviews=len(unpaid_five_star_reviews_df)
unpaid_five_star_reviews
```

```python
unpaid_reviews_ratio=round(unpaid_five_star_reviews/len(unpaid_reviews_df)*100,2)
unpaid_reviews_ratio
```
